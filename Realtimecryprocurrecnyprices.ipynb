```
price = get_crypto_price('bitcoin')

print(price)
```

```
def main():
  last_price = -1

  while True:
    crypt = 'bitcoin'
    price = get_crypto_price(crypto)
    if price != last_price:
      print(crypto+' price: ', price)
      last_price = price
    time.sleep(3)

```

```
main()
```
